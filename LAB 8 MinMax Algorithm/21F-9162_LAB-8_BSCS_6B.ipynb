```python
from collections import defaultdict

MIN,MAX=1000,-1000
tc,sc=0,0

graph={
    'A': ['B','C','D'],
    'B': ['E','F','G'],
    'C': ['H','I','J'],
    'D': ['K','L','M'],
    'E': [],
    'F': [],
    'G': [],
    'H': [],
    'I': [],
    'J': [],
    'K': [],
    'L': [],
    'M': []
}

cost = {
    'A':-1000,
    'B':1000,
    'C':1000,
    'D':1000,
    'E':3,
    'F':12,
    'G':18,
    'H':-1,
    'I':5,
    'J':8,
    'K':-10,
    'L':-5,
    'M':-3

}

def decision(graph,n):
    return max_val(graph,n)



def max_val(graph,n):
    global MAX
    global tc,sc
    sc+=1
    if graph[n] == []:
        return cost[n]
    
    if cost[n] > -1000:
        MAX = cost[n]
        return MAX
    
    for child in graph[n]:
        MAX = max( MAX ,min_val(graph,child))
        tc += 1
        cost[n] = MAX
     
    
    return MAX

def min_val(graph,n):
    global MIN
    global tc,sc
    sc+=1
    if graph[n] == []:
        return cost[n]
    
    if cost[n] < -1000:
        MIN = cost[n]
        return MIN
    
    for child in graph[n]:
        MIN = min( MIN ,max_val(graph,child))
        tc += 1
        cost[n] = MIN

    return MIN


print("FINAL COST: ",decision(graph,'A'))
print("TIME: ",tc)
print("SPACE: ",sc)
print("UPDATED COST: ",cost)


```
